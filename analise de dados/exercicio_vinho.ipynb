# **5 - Visualização de Dados:**
Vamos agora criar gráficos para responder às perguntas formuladas na fase de Definição do Problema.

**Pergunta 1:** *Quais variáveis têm maior correlação com a qualidade?*

```python
# Passo 1: Calcular a matriz de correlação usando apenas colunas numéricas
correlation_matrix = wine.select_dtypes(include='number').corr()

# Passo 2: Criar o heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Matriz de Correlação')
plt.show()
```

por meio deste gráfico, verificamos que as variáveis que mais apresentam correlação com a qualidade do vinho são: álcool, sulfatos e ácido cítrico

**vamos fazer um gráfico de correlação somente com essas variáveis**

```python
cols = ['alcohol','sulphates', 'citric acid', 'quality'] # quais colunas
plot_data = wine[cols]

# calculando a correlação entre as colunas selecionadas
corr = plot_data.corr()

# criando a figura
plt.figure(figsize=(10, 8))

# criando o heatmap com anotações dos valores de correlação
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, cbar=True)

# configurações adicionais do gráfico
plt.xticks(range(len(plot_data.columns)), plot_data.columns, rotation=45, ha='right')
plt.yticks(range(len(plot_data.columns)), plot_data.columns, rotation=0)
plt.title('Heatmap de Correlação com Valores Anotados')
plt.tight_layout()

plt.show()# exibir o gráfico
```

**Pergunta 2:** *Quais são as características mais comuns nos vinhos de alta e baixa qualidade?*

```python
# Criando uma coluna para a nova classificação de qualidade

# 7 ou superior: alta | 6 ou inferior: baixa

def definir_qualidade(numero_qualidade): # função
    if numero_qualidade >= 7:
        return 'Alta'
    else:
      return 'Baixa'

wine['qualidade'] = wine['quality'].apply(definir_qualidade)
# "aplicando a função" | "qualidade" é igual a "quality" mas aplicando a função "definir_qualidade" (feita ali em cima )

wine.head()
```

```python
# Separar os dados de qualidade (alta e baixa) em duas variáveis diferentes

vinhos_alta = wine[wine['qualidade'] == 'Alta']
vinhos_baixa = wine[wine['qualidade'] == 'Baixa']
```

```python
# Estatísticas descritivas para vinhos de alta qualidade

print("Estatísticas descritivas - Vinhos de alta qualidade")
print()
vinhos_alta.describe()
```

```python
# Estatísticas descritivas para vinhos de baixa qualidade

print("Estatísticas descritivas - vinhos de biaxa qualidade")
print()
vinhos_baixa.describe()
```

```python
# visualização das distribuições das variáveis em vinhos de diferentes qualidades
for column in wine.columns[:-2]:  # enão inclui 'quality' e 'qualidade'
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=wine, x='qualidade', y=column)
    plt.title(f'Distribuição de {column} por Categoria de Qualidade')
    plt.show()
```

**Pergunta 3:** *Quais subtâncias estão mais presentes nos vinhos?*

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=wine.drop(columns=['quality']))
plt.title("Distribuição das Substâncias no Vinho Tinto")
plt.ylabel("Concentração")
plt.xticks(rotation=30)
plt.show()
```

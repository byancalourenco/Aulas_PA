# **Estruturas condicionais**

Avaliar se o aluno foi **aprovado** caso sua média seja maior ou igual a 5

image.png

```python
# entradas das notas
n1 = float(input("Digite a primeira nota: "))
n2 = float(input("Digite a segunda nota: "))
n3 = float(input("Digite a terceira nota: "))
```

```python
# processamento da média
media = (n1 + n2 + n3) /3
media
```

```python
# avaliando o valor da média, se é maior ou igual a 5, o aluno será aprovado
if (media >= 5):
  print("Aprovado")
print("Fim do programa.....")
```

Avaliar se o aluno foi **`aprovadao ou reprovado`**, caso sua média seja maior ou igual a 5

image.png

```python
# entradas das notas
n1 = float(input("Digite a primeira nota: "))
n2 = float(input("Digite a segunda nota: "))
n3 = float(input("Digite a terceira nota: "))
```

```python
# processamento da média
media = (n1 + n2 + n3) /3
media
```

```python
# avaliando o valor da média, se é maior ou igual a 5
if (media >= 5):
  print("Aprovado")

else:
  print("Reprovado")

print("Fim do programa.....")
```

Avaliar se o aluno foi **`aprovado,reprovado ou está de recuperação`**, caso sua média seja:
aprovado media > 5,
reprovado media < 4,
recuperação media entre 4 e 5

image.png

```python
# entradas das notas
n1 = float(input("Digite a primeira nota: "))
n2 = float(input("Digite a segunda nota: "))
n3 = float(input("Digite a terceira nota: "))
```

```python
# processamento da média
media = (n1 + n2 + n3) /3
media
```

```python
# avaliando se está aprovado, reprovado ou de recuperação
if media>5:
  print("Aprovado")
else:
  if media<4:
    print("Reprovado")
  else:
    print("Recuperação")
```

```python
# outro jeito de fazer o passo anterior (abreviar o else e if: elif)
if media>5:
  print("Aprovado")
elif media<4:
  print("Reprovado")
else:
    print("Recuperação")
```

```python
# entradas das notas
n1 = float(input("Digite a primeira nota: "))
n2 = float(input("Digite a segunda nota: "))
n3 = float(input("Digite a terceira nota: "))
```

Avaliar se o aluno foi **`aprovadao ou reprovado`**, caso sua m√©dia seja maior ou igual a 5

```python
# entradas das notas
n1 = float(input("Digite a primeira nota: "))
n2 = float(input("Digite a segunda nota: "))
n3 = float(input("Digite a terceira nota: "))
```

Avaliar se o aluno foi **`aprovado,reprovado ou est√° de recupera√ß√£o`**, caso sua m√©dia seja:
aprovado media > 5,
reprovado media < 4,
recupera√ß√£o media entre 4 e 5

image.png

```python
# entradas das notas
n1 = float(input("Digite a primeira nota: "))
n2 = float(input("Digite a segunda nota: "))
n3 = float(input("Digite a terceira nota: "))
```

```python
# processamento da m√©dia
media = (n1 + n2 + n3) /3
media
```

```python
# avaliando se est√° aprovado, reprovado ou de recupera√ß√£o
if media>5:
  print("Aprovado")
else:
  if media<4:
    print("Reprovado")
  else:
    print("Recupera√ß√£o")
```

```python
# outro jeito de fazer o passo anterior (abreviar o else e if: elif)
if media>5:
  print("Aprovado")
elif media<4:
  print("Reprovado")
else:
    print("Recupera√ß√£o")
```
